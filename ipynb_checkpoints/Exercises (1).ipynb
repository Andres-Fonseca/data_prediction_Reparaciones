### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

# Cargar los datos en un DataFrame
df = pd.read_csv(url, sep='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df.copy()
```

### Step 4. How many products cost more than $10.00?

```python
def simbolo(x):
    """ funcion para quitar simbolo """
    
    return x.replace('$', '').strip()
def tipo(x):
    
    """ funcion para colocar tipo float """
    return float(x)
chipo['item_price'] = list(map(simbolo, chipo['item_price']))
chipo['item_price']= list(map(tipo, chipo['item_price']))
```

```python
print(f'hay {chipo[chipo.item_price>10.00].item_name.value_counts().count()} productos con precio mayot a 10')
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.groupby('item_name')['item_price'].min().reset_index()
```

### Step 6. Sort by the name of the item

```python
chipo.groupby('item_name')['item_price'].min().reset_index().sort_values(by='item_name', ascending=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
print(f' la cantidad del articulo mas vendido fue { chipo[['item_name','quantity']].groupby('item_name')['quantity'].sum().sort_values(ascending=False).reset_index().head(1)['quantity'].values[0]}')
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
print(f'el Veggie Salad Bowl se pidio { chipo[chipo.item_name=='Veggie Salad Bowl'].shape[0]} veces')
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
print(f'Canned Soda se vendio con mas de una cantidad en {chipo[(chipo.item_name=='Canned Soda') & (chipo.quantity>1)].shape[0]} veces')
```
